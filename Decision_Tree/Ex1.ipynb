## Motivating Random Forests: Decision Trees

隨機森林是建立在決策樹上的整體學習者(ensemble learner)的一個例子。 因此，我們將從討論決策樹本身開始。

決策樹是對對象進行分類或標記的非常直觀的方法：只需提出一系列旨在歸類到分類的問題。 <br>
例如，如果您想構建一棵決策樹以對遇到的動物進行分類，則可以構建以下的決策樹：<br>
Reference: https://github.com/jakevdp/PythonDataScienceHandbook
![Image](https://raw.githubusercontent.com/jakevdp/PythonDataScienceHandbook/8a34a4f653bdbdc01415a94dc20d4e9b97438965/notebooks/figures/05.08-decision-tree.png)

二進制拆分(binary splitting)決策樹極為有效：<br>
在結構良好的樹中，每個問題將使選項的數量減少大約一半，甚至在大量類中也非常迅速地縮小選項的範圍。

在決策樹的機器學習中，問題通常採取數據中軸對齊拆分的形式：也就是說，樹中的每個節點都使用其中一個特徵中的截止值(cutoff value)將數據拆分為兩組。

### Creating a decision tree
考慮以下二維數據，該數據具有四個類標籤之一：

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=300, centers=4,
                  random_state=0, cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

基於此數據的簡單決策樹將根據某種定量標準沿一個或另一個軸迭代地拆分數據，<br>
並在每個級別上根據新區域中的多數點投票為其分配標籤。
此圖顯示了此數據的決策樹分類器的前四個級別的視覺化： <br>
![image](https://raw.githubusercontent.com/jakevdp/PythonDataScienceHandbook/8a34a4f653bdbdc01415a94dc20d4e9b97438965/notebooks/figures/05.08-decision-tree-levels.png)

請注意，在第一次拆分之後，上部分支中的每個點均保持不變，因此無需進一步細分此分支。<br>
除了包含所有一種顏色的節點外，每個級別的"每個"區域(each level *every* region)都再次沿著兩個要素之一進行分割。

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier().fit(X, y)
```

視覺化分類器的輸出：

```python
def visualize_classifier(model, X, y, ax=None, cmap='rainbow'):
    ax = ax or plt.gca()
    
    # Plot the training points
    ax.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=cmap,
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # fit the estimator
    model.fit(X, y)
    xx, yy = np.meshgrid(np.linspace(*xlim, num=200),
                         np.linspace(*ylim, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    # Create a color plot with the results
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap=cmap, clim=(y.min(), y.max()),
                           zorder=1)

    ax.set(xlim=xlim, ylim=ylim)
```

現在我們可以檢查決策樹分類是什麼樣的：

```python
visualize_classifier(DecisionTreeClassifier(), X, y)
```

### Decision trees and over-fitting

請注意，隨著深度的增加，我們將得到形狀非常奇怪的分類區域。 <br>
例如，第五個的深度處，黃色和藍色區域之間有一個紫色區域。 <br>
顯然，這不是真正的分類，而是數據的特定採樣或噪音的結果。 <br>
也就是說，即使僅深五個級別的決策樹，顯然也會過度擬合(over-fitting)我們的數據。

一種組合方法(ensemble method)稱作*bagging*: 即是可以組合多個過度擬合的估計量以減少這種過度擬合的影響礎。 <br>
Baggin利用一組並行的估計量（也許是一個grab bag），每個估計量都過度擬合數據，並對結果求平均以找到更好的分類。<br>一組隨機決策樹被稱為隨機森林。

可以使用Scikit-Learn的BaggingClassifier估計器手動完成這種類型的分類，如下所示：

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier

tree = DecisionTreeClassifier()
bag = BaggingClassifier(tree, n_estimators=100, max_samples=0.8,
                        random_state=1)

bag.fit(X, y)
visualize_classifier(bag, X, y)
```

## Example: Random Forest for Classifying Digits

我們將使用手寫數字數據，來看如何在這種情況下使用隨機森林分類器。

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.keys()
```

視覺化前幾個數據，來更好地了解資料

```python
# set up the figure
fig = plt.figure(figsize=(6, 6))  # figure size in inches
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)

# plot the digits: each image is 8x8 pixels
for i in range(64):
    ax = fig.add_subplot(8, 8, i + 1, xticks=[], yticks=[])
    ax.imshow(digits.images[i], cmap=plt.cm.binary, interpolation='nearest')
    
    # label the image with the target value
    ax.text(0, 7, str(digits.target[i]))
```

我們可以使用隨機森林對數字進行快速分類，如下所示：

```python
from sklearn.model_selection import train_test_split

Xtrain, Xtest, ytrain, ytest = train_test_split(digits.data, digits.target,
                                                random_state=0)
model = RandomForestClassifier(n_estimators=1000)
model.fit(Xtrain, ytrain)
ypred = model.predict(Xtest)
```

我們可以透過查看此分類器的分類報告，來評估分類結果是否準確：

```python
from sklearn import metrics
print(metrics.classification_report(ypred, ytest))
```

為了更好地度量，繪製混淆矩陣(confusion matrix)：

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(ytest, ypred)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

我們發現，簡單的、未經調整的隨機森林可對數字數據進行非常準確的分類。

## Summary of Random Forests

我們介紹了隨機森林(一組隨機決策樹)的概念。
隨機森林是一種強大的方法，具有以下優點：

-由於底層決策樹的簡單性，訓練和預測都非常快。另外，這兩個任務可以並行化，因為各個樹是完全獨立的實體。
-多棵樹可以進行機率分類：估計數中的多數票給出了機率的估計值。
-非參數模型非常靈活，因此可以很好地處理其他估算器無法滿足的任務。

隨機森林的一個主要缺點是結果不容易解釋：也就是說，如果您想得出有關分類模型的*含義*的結論，則隨機森林可能不是最佳選擇。
