# In Depth: k-Means Clustering

我們將介紹一種無監督的機器學習模型：聚類算法(Clustering Algorithm)。
聚類算法試圖從數據的屬性(properties)中學習點組的最佳劃分(optimal division)或離散標記(discrete labeling of groups of points)。

Scikit-Learn和其他地方提供了許多聚類算法，但也許最容易理解的是稱為"k-means聚類"的算法，該算法在sklearn.cluster.KMeans中實現。

Reference: https://github.com/jakevdp/PythonDataScienceHandbook

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
```

## Introducing k-Means

k-means算法在未標記的多維數據集中搜索預定數量的聚類(pre-determined number of clusters)。<br>
使用聚類方法的目的為：<br><br>

-"聚類中心"(cluster center)是屬於群集的所有點的算術平均值。 <br>
-每個點比其他聚類中心(cluster center)更靠近自己的聚類中心。 <br>

以上的兩個假設是k-means模型的基礎。<br>
我們很快將深入研究算法如何達到此解決方案，但是現在讓我們看一下簡單的數據集並查看k均值結果。<br>

首先，讓我們生成一個包含四個不同Blob的二維數據集。<br>
為了強調這是一種無監督算法，我們將標籤從可視化中排除。<br>

註: 群集 = 集群 = 聚類 = 叢，以上這幾個名詞在不同解釋k-means的文本中皆有提到，但都為同樣意思。

```python
from sklearn.datasets.samples_generator import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

我們可以用肉眼觀察到四個聚類。 <br>
而k-means算法即是自動執行此操作(分成四群)，在Scikit-Learn中使用典型的估算器(estimator) API：

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

我們通過繪製由這些標籤著色的數據來視覺化結果。 <br>
並繪製由k-means估計器確定的聚類中心(黑色圓型)： 

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

聚類分配的可能組合數量在數據點數量上成指數增長，而詳盡的搜索將非常耗時。<br>
而對我們來說幸運的是，這樣的詳盡搜索是不必要的： <br>
k-means的典型方法涉及一種稱為* expectation-maximization *的直觀迭代方法。

## k-Means Algorithm: Expectation–Maximization

Expectation–maximization (E–M) 是一種功能強大的算法，可用於數據科學中的各種情況。<br>
k-means 是該算法的一個特別簡單且易於理解的應用，我們將在此簡要地介紹它。<br>
簡而言之，此處的期望最大化方法包括以下過程：<br><br>

1.隨機猜測一些聚類中心<br>
2.重複直到收斂<br>
   2.1. * E-Step *：將點分配給最近的聚類中心
   2.2. * M-Step *：將聚類中心設置為均值
   
這裡的"E-step" 或是 "Expectation step"之所以如此命名是因為它涉及更新我們對每個點所屬的聚類的期望。<br><br>
而"M-step" 或是 "Maximization step"，是因為它涉及最大化一些定義聚類中心位置的適應度函數(fitness function)，<br>在這種情況下，最大化是通過對每個聚類中的數據取平均值(mean)來實現的。 

關於該算法的paper很多，但可以總結如下：<br>
在典型情況下，E-step和M-step的每次重複總是會更好地估計聚類特徵。<br>
我們可以將算法視覺化，如下圖所示。

![img](https://github.com/jakevdp/PythonDataScienceHandbook/raw/8a34a4f653bdbdc01415a94dc20d4e9b97438965/notebooks/figures/05.11-expectation-maximization.png)

 k-Means算法非常簡單，我們可以用幾行代碼完成它。
以下是一個非常基本的實現：

```python
from sklearn.metrics import pairwise_distances_argmin

def find_clusters(X, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
    centers = X[i]
    
    while True:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)
        
        # 2b. Find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels

centers, labels = find_clusters(X, 4)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

### Caveats of expectation–maximization

使用期望最大化算法時，需要注意一些問題。

#### The globally optimal result may not be achieved

首先，儘管可以保證通過E-M steps可以改善每個步驟的結果，但不能保證它會導致全域(global)最佳解決方案。
例如，如果我們在簡單的過程中使用其他隨機種子(random seed)，則特定的開始猜測(particular starting guesses)會導致較差的結果：

```python
centers, labels = find_clusters(X, 4, rseed=0)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

此處，E-M steps已收斂(converged )，但尚未收斂到全域最佳配置(globally optimal configuration)。 <br>
由於這個原因，通常會多次嘗試運行該算法，就像Scikit-Learn實際上默認情況下所做的那樣（由n_init參數設置，默認為10）。

#### The number of clusters must be selected beforehand
k-means算法 另一個常見挑戰是，您必須告訴它期望有多少個群集：它無法從數據中了解群集的數量。 <br>
例如，如果我們要求算法識別六個聚類，它將進行並找到最佳的六個聚類：

```python
labels = KMeans(6, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

或者，您可以使用更複雜的聚類算法，該算法對每個聚類數量的適應性具有更好的定量度量<br>
（例如，Gaussian mixture models ...);<br>
或 選擇合適數量的集群（例如，DBSCAN, mean-shift, affinity propagation）)

#### k-means is limited to linear cluster boundaries

k-menas 群集之間的邊界(boundaries)將始終是線性的，這意味著它將無法處理更複雜的邊界。 <br>
考慮以下數據，以及k-means方法發現的群集標籤：

```python
from sklearn.datasets import make_moons
X, y = make_moons(200, noise=.05, random_state=0)
```

```python
labels = KMeans(2, random_state=0).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

這種情況使人聯想到(SVM, Support-Vector-Machines），<br> 
在SVM中，我們使用內核(kernal)變換將數據投影到可以進行線性分離的更高維度。<br>
我們可能會想使用相同的技巧允許k-means發現非線性邊界。<br>

在Scikit-Learn中的"SpectralClustering" 估算器(estimator)中實現了<br>
該內核化(kernelized ) k-means的一個版本。<br>
它使用最近鄰居(nearest neighbors)的圖來計算數據的高維表示，然後使用k-means算法分配標籤：

```python
from sklearn.cluster import SpectralClustering
model = SpectralClustering(n_clusters=2, affinity='nearest_neighbors',
                           assign_labels='kmeans')
labels = model.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=50, cmap='viridis');
```

我們看到，使用這種方法，被內核化的* k * -means能夠找到簇之間更複雜的非線性邊界。

## Examples

注意這些算法的局限性，我們可以在各種情況下使用 k-means來發揮優勢。
以下觀察幾個範例:

### Example 1: k-means on digits

首先，讓我們看一下對在decision_tree_example中看到的相同數字數據。 <br>
我們將嘗試使用k-means來嘗試識別類似的數字，而無需使用原始標籤訊息。<br>
(這似於從沒有任何先驗標籤訊息的新數據集中提取的第一步)<br>

註: 這些數字由具有64個特徵的1,797個樣本組成，其中64個特徵中的每一個都是8×8圖像中一個像素的亮度：

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

可以像以前一樣執行聚類：

```python
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

結果是在64個維度上有10個簇。<br>
請注意，聚類中心本身是64維點(64-dimensional point)，它們本身可以解釋為聚類中的典型數字(digit)。<br>
讓我們看看這些集群中心是什麼樣的： 

```python
fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

我們看到即使沒有標籤，K-means也能夠找到其中心是可識別數字的聚類，也許是1和8除外。<br><br>

由於k-means對群集的身份一無所知，因此0–9標籤可能會被置換。<br>
所以我們可以通過將每個學習到的集群標籤與在其中找到的真實標籤進行配對來解決此問題：

```python
from scipy.stats import mode

labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]
```

現在，我們可以檢查我們的無監督聚類在數據中查找相似數字時的準確性：

```python
from sklearn.metrics import accuracy_score
accuracy_score(digits.target, labels)
```

僅需簡單的k-means算法，我們就為80％的輸入數字找到了正確的分類！ <br>
(註: 無監督學習的準確率通常較低)<br>
讓我們檢查一下混淆矩陣：

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(digits.target, labels)
sns.heatmap(mat.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

正如我們可能從以前看到的聚類中心所期望的那樣，混淆的主要點是8與1之間。<br>
但這仍然表明，使用k-means，我們基本上可以構建數字分類器，而無需引用任何已知標籤。<br><br>

為了比較，讓我們嘗試進一步測試。<br>
我們可以使用t分佈隨機鄰居嵌入（t-SNE）算法對數據進行預處理，然後再執行k-means 。<br>
t-SNE是一種非線性嵌入算法，特別擅長於保留群集中的點。<br>
讓我們看看它是如何工作的：

```python
from sklearn.manifold import TSNE

# Project the data: this step will take several seconds
tsne = TSNE(n_components=2, init='random', random_state=0)
digits_proj = tsne.fit_transform(digits.data)

# Compute the clusters
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(digits_proj)

# Permute the labels
labels = np.zeros_like(clusters)
for i in range(10):
    mask = (clusters == i)
    labels[mask] = mode(digits.target[mask])[0]

# Compute the accuracy
accuracy_score(digits.target, labels)
```

不使用標籤情況下，分類準確率接近92％。 <br>
如果使用的好，這就是無監督學習的能力：它可以從數據集中提取可能很難用手工或肉眼觀察到的信息。

### Example 2: *k*-means for color compression

聚類的一種有趣的應用是圖像內的顏色壓縮。<br>
例如，假設有一幅具有數百萬種顏色的圖像。<br>
在大多數圖像中，將不使用大量顏色，並且圖像中的許多像素將具有相似甚至相同的顏色。。<br>

```python
# Note: this requires the ``pillow`` package to be installed
from sklearn.datasets import load_sample_image
china = load_sample_image("china.jpg")
ax = plt.axes(xticks=[], yticks=[])
ax.imshow(china);
```

圖像本身存儲在尺寸為(height，width，RGB)的三維數組中，其中包含紅色/藍色/綠色，為0到255之間的整數：

```python
china.shape
```

我們可以查看這組像素的一種方法是在三維顏色空間中將點雲包圍( a cloud of points )。 <br>
我們將數據重塑為 ``[n_samples x n_features]``，並重新縮放顏色，使其位於0到1之間：

```python
data = china / 255.0 # use 0...1 scale
data = data.reshape(427 * 640, 3)
data.shape
```

我們可以使用10,000個像素的子集來可視化此顏色空間中的這些像素：

```python
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data
    
    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T
    
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20);
```

```python
plot_pixels(data, title='Input color space: 16 million possible colors')
```

現在，使用跨整個像素空間的k-means聚類方法，將這1600萬種顏色減少為16種顏色。<br>
因為我們要處理非常大的數據集，所以我們將使用迷你批(mini-batch)處理k-means，<br>
批處理對數據的子集進行運算，比標準k-means算法更快地計算出結果：

```python
import warnings; warnings.simplefilter('ignore')  # Fix NumPy issues.

from sklearn.cluster import MiniBatchKMeans
kmeans = MiniBatchKMeans(16)
kmeans.fit(data)
new_colors = kmeans.cluster_centers_[kmeans.predict(data)]

plot_pixels(data, colors=new_colors,
            title="Reduced color space: 16 colors")
```

結果是對原始像素進行重新著色，其中為每個像素分配了其最接近的聚類中心的顏色。<br>
在圖像空間而不是像素空間中繪製這些新顏色將向我們展示這種效果：

```python
china_recolored = new_colors.reshape(china.shape)

fig, ax = plt.subplots(1, 2, figsize=(16, 6),
                       subplot_kw=dict(xticks=[], yticks=[]))
fig.subplots_adjust(wspace=0.05)
ax[0].imshow(china)
ax[0].set_title('Original Image', size=16)
ax[1].imshow(china_recolored)
ax[1].set_title('16-color Image', size=16);
```

在最右邊的面板中肯定會丟失一些細節，但是整體圖像仍然很容易辨認。<br>
右邊的這張圖實現了大約一百萬的壓縮係數！<br>
儘管這是k-means的有趣應用，但是肯定有更好的方法來壓縮圖像中的信息。<br>
